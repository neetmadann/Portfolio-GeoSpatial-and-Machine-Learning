### Code to grab the data 

```python
import pandas as pd
import altair as alt
import os
os. getcwd()
path = r'C:\Users\vaibh\Downloads\CW Group 2\CW Group 2'
os.chdir(path)
df = pd.read_csv('gapminder.csv')
df.head()
len(df)
```

## Data Cleaning

```python
global_df = df.groupby(['Year','region']).mean().reset_index()
pd.options.display.float_format = '{:20}'.format
global_df.head()
```

```python
brush = alt.selection_interval()  # selection of type "interval"
chart = alt.Chart(global_df).mark_line().encode(
    alt.X('Year:Q',axis=alt.Axis(format="")), #change the year take out comma
    color=alt.condition(brush, 'region:N', alt.value('lightgray')
    
    )
).properties(
    width=450,
    height=300
).add_selection(
    brush
)

chart.encode(y='gdp:Q')
```

```python
brush = alt.selection_interval()  # selection of type "interval"
chart = alt.Chart(global_df).mark_line().encode(
    alt.X('Year:Q',axis=alt.Axis(format="")), #change year take out comma
    color=alt.condition(brush, 'region:N', alt.value('lightgray'))
).properties(
    width=300,
    height=200
).add_selection(
    brush
)
chart.encode(y='life:Q')|chart.encode(y='child_mortality:Q')|chart.encode(y='fertility:Q').interactive()
```

```python
#RELAtionship
gapminder_2013 = df[df['Year']==2013]

alt.Chart(gapminder_2013).mark_circle().encode(
    alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative'),
    color='region:N'
).properties(
    width=150,
    height=150
).repeat(
    row=['fertility','life','population','child_mortality','gdp'],
    column=['fertility','life','population','child_mortality','gdp']
).interactive()
```

```python
df.region.unique()
region_df = global_df
region_df['gdp_per_100']=region_df['gdp']/100
region_df = pd.melt(region_df,id_vars=['Year','region'],value_vars=['gdp_per_100','fertility','child_mortality','life'])
region_df.head()
```

```python
sort_des = global_df.groupby(['region']).sum().reset_index()
sort_des = sort_des.sort_values(by='gdp', ascending=False)
sort_des
```

```python
region_list = list(sort_des['region'].unique())
high_region = region_list[0:4]
high_region
low_region=region_list[-2:]
low_region
```

```python
region_df_expand = global_df
region_df_expand['gdp_per_1000'] = global_df['gdp']/1000
region_df_expand['pop_per_mil'] = global_df['population']/1000000
region_df_expand = pd.melt(global_df,id_vars=['Year','region'],value_vars=['fertility','child_mortality','life'])
#'gdp_per_1000','pop_per_mil',
```

```python
from altair.expr import datum

print('\033[1m' + 'HIGH GDP COUNTRIES' + '\033[0m')

#'Sub-Saharan Africa',
chart = alt.hconcat()
for region in high_region:
    base = alt.Chart(region_df,title=region).mark_line().encode(
    alt.X('Year:O',title='Year',axis=alt.Axis(values=list(range(1,50,10)))),
    alt.Y('value:Q',title='Socio-Economic Measures'),
    alt.Color('variable:N')
).interactive().properties(
    width=200,
    height=200
)
    chart |= base.transform_filter(datum.region == region)
chart
```

```python
from altair.expr import datum

print('\033[1m' + 'FERTILITY,MORTALITY,LIFE' + '\033[0m')

#'Sub-Saharan Africa',
chart = alt.hconcat()
for region in high_region:
    base = alt.Chart(region_df_expand,title=region).mark_line().encode(
    alt.X('Year:O',title='Year',axis=alt.Axis(values=list(range(1,50,10)))),
    alt.Y('value:Q',title='Socio-Economic Measures'),
    alt.Color('variable:N')
).interactive().properties(
    width=200,
    height=200
)
    chart |= base.transform_filter(datum.region == region)
chart
```
